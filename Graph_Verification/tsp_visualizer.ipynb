- fi10639 = https://www.math.uwaterloo.ca/tsp/world/fitour.html 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def read_tsp_file(filename):
    """Reads city coordinates from a TSP file."""
    cities = []
    with open(filename, 'r') as file:
        lines = file.readlines()
        cities_start = lines.index('NODE_COORD_SECTION\n') + 1
        cities_end = lines.index('EOF\n')
        cities_data = lines[cities_start:cities_end]
        for line in cities_data:
            parts = line.split()
            if len(parts) >= 3:
                _, x, y = parts
                cities.append([float(x), float(y)])
    return np.array(cities)

def read_tour_data_from_file(filename):
    """Reads the best tour, tour length, and time taken from a file."""
    best_tour = []
    tour_length = 0.0
    time_taken = 0.0
    with open(filename, 'r') as file:
        lines = file.readlines()
        for line in lines:
            if line.startswith("Best tour:"):
                best_tour = list(map(int, lines[lines.index(line) + 1].strip().split()))
            elif line.startswith("Tour length:"):
                tour_length = float(lines[lines.index(line) + 1].strip())
            elif line.startswith("Time taken:"):
                time_taken = float(lines[lines.index(line) + 1].strip())
    return best_tour, tour_length, time_taken

def plot_tsp_solution(cities, tour):
    """Plots the TSP solution given the city coordinates and tour."""
    plt.figure(figsize=(12, 8))
    plt.scatter(cities[:, 0], cities[:, 1], color='blue', marker='o', s=0, label='Cities')  # Smaller dots with s=10
    tour_cities = np.append(tour, tour[0])  # To close the tour loop
    plt.plot(cities[tour_cities, 0], cities[tour_cities, 1], color='red', linestyle='-', marker='o', markersize=0, linewidth=1, label='Tour Path')  # Thinner line and smaller red dots with markersize=4
    plt.title('TSP Solution')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.legend()
    plt.grid(True)
    plt.show()

st_tour_length = []
st_time_taken = []
omp_tour_length = []
omp_time_taken = []
cuda_tour_length = []
cuda_time_taken = []
```

# **SERIAL CODE**

## qa194

```python
# Example usage
cities = read_tsp_file('data/qa194.tsp')
best_tour, tour_length, time_taken = read_tour_data_from_file('st_result/qa194.txt')
st_tour_length.append(tour_length)
st_time_taken.append(time_taken)
plot_tsp_solution(cities, best_tour)

# Optionally, print tour details
print(f"Tour length: {tour_length}")
print(f"Time taken: {time_taken} seconds")
```

# **TIME COMPARISON**

```python
import matplotlib.pyplot as plt

def plot_time_comparison(st_time_taken, omp_time_taken, cuda_time_taken):
    """Plots a comparison of the time taken by different TSP solvers."""
    
    # X-axis labels
    labels = ['qa194', 'pbm436', 'lu980', 'tz6117', 'fi10639']
    
    # Plotting the time taken for each method with custom colors and markers
    plt.figure(figsize=(10, 6))
    plt.plot(labels, st_time_taken, label='Single-threaded', marker='o', color='blue', markerfacecolor='blue')
    plt.plot(labels, omp_time_taken, label='OpenMP', marker='s', color='green', markerfacecolor='green')
    plt.plot(labels, cuda_time_taken, label='CUDA', marker='D', color='red', markerfacecolor='red')
    
    # Adding titles and labels
    plt.title('Time Comparison for Different TSP Solvers')
    plt.xlabel('Dataset')
    plt.ylabel('Time Taken (s)')
    plt.legend()
    plt.grid(True)
    
    # Show the plot
    plt.show()

# Example usage:
# Assuming st_time_taken, omp_time_taken, and cuda_time_taken are defined
plot_time_comparison(st_time_taken, omp_time_taken, cuda_time_taken)
```
